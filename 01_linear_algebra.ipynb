### When will a bank robber be caught

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
time_sequence = np.linspace(start=0, stop=40, num=1000)
```

```python
distance_travelled_by_robber = 2.5 * time_sequence 
distance_travelled_by_sheriff = 3.0 * (time_sequence -5)
```

```python
fig, ax = plt.subplots()

plt.title("When does the bank robber get caught")
plt.xlabel("time (in minutes)")
plt.ylabel("distance (in km)")

ax.set_xlim([0,40])
ax.set_ylim([0,100])

ax.plot(time_sequence, distance_travelled_by_robber, color='red')
ax.plot(time_sequence, distance_travelled_by_sheriff, color='green')

plt.axvline(x=30,color='purple', linestyle='--')
plt.axhline(y=75,color='purple', linestyle='--')
```

### Solar Power

mark_1_energy = 1*t <br>
mark_2_energy = 4(t-30) - as it starts a month later


```python
time_sequence = np.linspace(start=0, stop=50, num=100)
```

```python
mark_1_energy_sequence = time_sequence
mark_2_energy_sequence = 4*(time_sequence-30)
```

```python
fig, ax = plt.subplots()

plt.title("when does Mark 2 generate the same energy as the Mark 1")
plt.xlabel("time (in minutes)")
plt.ylabel("Energy (in kJ)")

ax.set_xlim([0,50])
ax.set_ylim([0,100])

ax.plot(time_sequence, mark_1_energy_sequence, color='brown')
ax.plot(time_sequence, mark_2_energy_sequence, color='purple')
```
